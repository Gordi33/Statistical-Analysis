User-Interactive Ticker Selection

```python
import ipywidgets as widgets
from IPython.display import display

user_selected = {}

def tickerSelection():
    checkboxes = {name: widgets.Checkbox(value=False, description=name) for name in tickers.keys()}
    def on_submit_clicked(b):
        global user_selected
        user_selected = {name: tickers[name] for name, cb in checkboxes.items() if cb.value}
        print("You selected:")
        for name, ticker in user_selected.items():
            print(f"  {name}: {ticker}")

    submit_button = widgets.Button(description="Submit Selection")
    submit_button.on_click(on_submit_clicked)

    # Display all checkboxes and submit button
    for cb in checkboxes.values():
        display(cb)
    display(submit_button)

tickerSelection()
```

```python
def add_indices_interactively(user_dict):
    name_input = widgets.Text(description="ETF Name:")
    ticker_input = widgets.Text(description="Ticker:")
    add_button = widgets.Button(description="Add ETF")
    output = widgets.Output()

    def on_add_clicked(b):
        with output:
            output.clear_output()
            name = name_input.value.strip()
            ticker = ticker_input.value.strip()
            if not name or not ticker:
                print("Please provide both name and ticker!")
                return
            if name in user_dict:
                print(f"ETF '{name}' already exists.")
                return
            user_dict[name] = ticker
            print(f"Added ETF:\n  {name}: {ticker}")
            print("\nCurrent selection:")
            for k, v in user_dict.items():
                print(f"  {k}: {v}")
            # Clear inputs for next add
            name_input.value = ""
            ticker_input.value = ""

    add_button.on_click(on_add_clicked)
    display(name_input, ticker_input, add_button, output)

# Usage:
add_indices_interactively(user_selected)
```

Plot Selected List

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

def plotPercentageGrowthWithPlotly():

    start_date = "2015-08-10"
    end_date = "2025-08-10"

    combined = pd.DataFrame()

    for name, ticker in user_selected.items():
        df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True)
        if df.empty:
            print(f"No data for {ticker} - skipping")
            continue
        combined[name] = df['Close']

    if combined.empty:
        raise ValueError("No data downloaded for any ticker")

    # Calculate percentage growth since start date
    pct_growth = combined.div(combined.iloc[0]).sub(1).mul(100)

    df_long = pct_growth.reset_index().melt(id_vars="Date", var_name="ETF", value_name="Pct Growth")

    # Plot interactive chart
    fig = px.line(df_long, x="Date", y="Pct Growth", color="ETF",
                  title="Percentage Growth of Selected ETFs Since Start Date",
                  labels={"Pct Growth": "Percentage Growth (%)"})

    fig.update_layout(legend_title_text='Click legend items to toggle visibility')

    fig.show()

plotPercentageGrowthWithPlotly()
```
